Buatlah eksperimen memakai data Parkinson (dimensi : 22, class: PD atau Healthy). Gunakan 100 sampel pertama sebagai training set dan sisanya (95 sampel) sebagai testing set. 

```python
import pandas as pd

file_path = 'parkinson.csv'
data = pd.read_csv(file_path)

data
```

```python
data.head(), data.info()
```

## KNN

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

X = data.drop(columns=['Class'])
y = data['Class']

X_train, X_test = X[:100], X[100:]
y_train, y_test = y[:100], y[100:]

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
report = classification_report(y_test, y_pred, target_names=['Healthy', 'PD'])
accuracy = accuracy_score(y_test, y_pred)
print(report)
print(f'Akurasi KNN: {accuracy}')
```

Model KNN mencapai akurasi sekitar 64.2%, yang berarti model ini berhasil mengklasifikasikan sekitar 64% dari data uji dengan benar. Namun, akurasi ini masih dapat dikatakan rendah untuk aplikasi medis yang umumnya memerlukan akurasi yang lebih tinggi.

## NAIVE BAYES

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
report_nb = classification_report(y_test, y_pred_nb, target_names=['Healthy', 'PD'])
print(report_nb)
print(f'Akurasi Naive Bayes: {accuracy_nb}')
```

Naive Bayes hanya mencapai akurasi sekitar 49.5%, yang lebih rendah dibandingkan dengan KNN (64.2%). Ini menunjukkan bahwa Naive Bayes kurang mampu mengidentifikasi pola pada data ini dibandingkan KNN. Dengan akurasi di bawah 50%.

## DECISSON TREE

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

dt = DecisionTreeClassifier(random_state=0)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
report_dt = classification_report(y_test, y_pred_dt, target_names=['Healthy', 'PD'])
print(report_dt)
print(f'Akurasi Decision Tree: {accuracy_dt}')
```

Dengan akurasi 37.9%, Decision Tree menunjukkan performa yang lebih rendah dibandingkan KNN (64.2%) dan Naive Bayes (49.5%). Akurasi ini menunjukkan bahwa Decision Tree kurang mampu mengklasifikasikan data Parkinson dengan benar.

## RANDOM FORREST

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
report_rf = classification_report(y_test, y_pred_rf, target_names=['Healthy', 'PD'])
print(report_rf)
print(f'Akurasi Random Forest: {accuracy_rf}')
```



Dari hasil eksperimen dengan beberapa algoritma machine learning untuk mendeteksi Parkinson menggunakan data yang diberikan, diperoleh hasil akurasi sebagai berikut:

1. **Random Forest**: 61.05%
2. **K-Nearest Neighbors (KNN)**: 64.21%
3. **Naive Bayes**: 49.47%
4. **Decision Tree**: 37.89%

### Kesimpulan

- **KNN** menghasilkan akurasi tertinggi (64.21%) dibandingkan algoritma lain. Hal ini menunjukkan bahwa metode berbasis tetangga terdekat cukup efektif dalam mengenali pola data pasien Parkinson dan membedakan antara pasien yang sehat dan yang memiliki Parkinson.
- **Random Forest** menunjukkan performa yang baik dengan akurasi 61.05%, menandakan bahwa ensemble learning juga memberikan klasifikasi yang cukup kuat. Namun, performanya masih di bawah KNN dalam eksperimen ini.
- **Naive Bayes** dan **Decision Tree** memiliki akurasi yang lebih rendah, menunjukkan bahwa pendekatan berbasis probabilistik sederhana dan pohon keputusan tunggal mungkin tidak cukup kuat dalam menangani kompleksitas dataset ini.
- **Imbalance** dalam akurasi deteksi antara kelas "Healthy" dan "PD" pada beberapa algoritma (khususnya Random Forest) menunjukkan bahwa model dapat lebih sulit dalam mendeteksi kelas "PD". Ini bisa menjadi indikasi bahwa dataset memerlukan lebih banyak penyeimbangan kelas atau fitur tambahan yang lebih jelas memisahkan kedua kelas.

Secara keseluruhan, KNN adalah model terbaik dalam eksperimen ini, tetapi perlu dipertimbangkan lebih lanjut metode lain atau tuning hyperparameter untuk meningkatkan akurasi dan konsistensi dalam mendeteksi penyakit Parkinson.
